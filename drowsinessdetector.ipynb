# **Creating a YAML file for YOLOv8 configurations**

```python
import yaml

data_yaml = dict(
    train ='Driver-drowsiness-detection-1/train',
    val ='Driver-drowsiness-detection-1/valid',
    test='Driver-drowsiness-detection-1/test',
    nc = 2,
    names =['awake','drowsy']
)

with open('data.yaml', 'w') as outfile:
    yaml.dump(data_yaml, outfile, default_flow_style=True)
    
%cat data.yaml
```

# **Training the YOLOv8 Large model**

```python
from ultralytics import YOLO
model = YOLO('yolov8l.pt')
model.train(data='/kaggle/working/data.yaml', epochs=50, imgsz=640, batch=16,exist_ok=True)
```
